```python
#sort by team and date
fdsorted = lenght.sort_values(['HomeTeam', 'gmDateTime'], ascending=[True, True])
fdsorted = fdsorted.reset_index() 

#get index range for each team
coord = []
for i in range(1,len(fdsorted)-1):
    if i == 1:
        x = [fdsorted.loc[0,'HomeTeam'],0]
    if fdsorted.loc[i,'HomeTeam'] != fdsorted.loc[i+1,'HomeTeam']:
        x.append(i)
        coord.append(x)
        x = [fdsorted.loc[i+1,'HomeTeam'],i+1]
    if i == (len(fdsorted)-2):
        x.append(i+1)
        coord.append(x)
```

```python
#variables to move to previous game
variaveis_subir = ['teamconfLoss','teamconfWin','teamawayLoss','teamawayWin','teamstkType','teamstkTot',
                   'teamgameBack','teamgameLost','teamgamePlay','teamgameWon','teamhomeLoss','teamhomeWin',
                   'teamlastFive','teamlastTen','teamptsAgnst','teamptsAllow','teamptsDiff','teamptsFor',
                   'teamptsScore','teamrank','teamrankOrd']

for var in variaveis_subir:
    for i in coord:
        inicio = i[1]
        fim = i[2]
        for i in range(fim,inicio+1,-1):
            fdsorted.loc[i,var] = fdsorted.loc[i-1,var]
        if var == 'teamstkType' or var == 'teamrankOrd':
            fdsorted.loc[inicio,var] = "None"
        else: 
            fdsorted.loc[inicio,var] = 0 
            
#variables to calculate last 5 game average
variaveis_media = ['team2PA','teamSTL','teamOrtg','teamSTL%','teamTS%', 'teamORB','teamTREB%', 'teamSTL/TO',
                   'teamTO','teamTRB', 'teamPF','team2PM','team2P%','team3P%','team3PA','team3PM','teamAR',
                   'teamASST%','teamAST','teamAST/TO','teamBLK','teamBLK%','teamBLKR','teamDRB','teamDREB%',
                   'teamDayOff','teamDrtg','teamEDiff','teamEFG%','teamFG%','teamFGA','teamFGM','teamFT%',
                   'teamFTA','teamFTM','teamptsAgnst','teamptsAllow','teamptsDiff','teamptsFor','teamptsScore','teamPTS']

for var in variaveis_media:
    for i in coord:
        inicio = i[1]
        fim = i[2]
        media = fdsorted.loc[inicio:fim,var].mean()
  
        for i in range(fim,inicio+4,-1):
            fdsorted.loc[i,var] = fdsorted.loc[i-5:i-1,var].mean()
        for i in range(inicio+4,inicio,-1):
            y = (i-inicio)
            values = list(fdsorted.loc[i-y:i-1,var])
            values.append(media)
            fdsorted.loc[i,var] = (sum(values) / len(values) )
        fdsorted.loc[inicio,var] = media
        
```

```python
#get variables we didnt edit
fdfl = list(fdsorted)
edited = variaveis_media+variaveis_subir
save = list(set(fdfl) - set(edited))
#remove those which much stay
save.remove('gmDateTime')
save.remove('teamRslt')
save.remove('index')
save.remove('offNm1')
save.remove('offNm2')
save.remove('offNm3')
save.remove('teamAbbr')
save.remove('teamDiv')
save.remove('away')
save.remove('HomeTeam')
#drop remaining
fdsorted= fdsorted.drop(columns=save)
#create two copies : one for oppt and one for home team
completeoppt = fdsorted.copy()
completeoppt = completeoppt[completeoppt['away'] == 'away']
completeteam = fdsorted.copy()
completeteam = completeteam[completeteam['away'] != 'away']
completeoppt = completeoppt.rename(columns={'HomeTeam': 'AwayTeam'})
#rename oppt columns
import re
x = list(completeoppt)
for i in range(0,len(x)):
    if 'team' in x[i]:
        y = re.sub('team', '', x[i])
        y = 'oppt' + y
        x[i] = y
completeoppt.columns = x

#drop signal columns
completeoppt = completeoppt.drop('away', axis = 1)
completeteam = completeteam.drop('away', axis = 1)

#get common variables to merge on
opptlist = list(completeoppt)
teamlist = list(completeteam)
x = list(set(opptlist) & set(teamlist))
#merge into a dataframe with one single game per row, with both home team and oppt team stats
finaldf= pd.merge(completeteam,completeoppt, on=x, how='inner')
#drop more redudant variables
finaldf = finaldf.drop('teamAbbr', axis = 1)
finaldf = finaldf.drop('opptAbbr', axis = 1)

finaldf = finaldf.drop('gmDateTime', axis = 1)
#reset original index
finaldf = finaldf.sort_values('index', ascending=True)
finaldf = finaldf.set_index('index')
```

## Multi-layer Perceptron 

```python
from sklearn.neural_network import MLPClassifier
#dictionary to store results
MLPCopt = {}
#Lists of parameters to test
act = ['identity', 'logistic', 'tanh', 'relu']
sol = ['lbfgs', 'sgd', 'adam']
lr = ['constant', 'invscaling', 'adaptive']
#iterate over each possible combination
for activator in act:
    for solver in sol:
        if solver == 'sgd' or 'adam':
            shuffle = False
        if solver == 'sgd':
            for rate in lr:
                #build model
                MLPC = MLPClassifier(solver=solver, activation=activator, learning_rate=rate,random_state=0).fit(dummyX_train, dummyy_train)
                #make predictions
                MLPCpreds = MLPC.predict(dummyX_test)
                comb = activator + solver + rate
                #save accuracy result
                MLPCopt[comb] = accuracy_score(dummyy_test, MLPCpreds)
        else: 
            #build model
            MLPC = MLPClassifier(solver=solver, activation=activator,random_state=0).fit(dummyX_train, dummyy_train)
            #make predictions
            MLPCpreds = MLPC.predict(dummyX_test)
            comb = activator + solver
            #save accuracy result
            MLPCopt[comb] = accuracy_score(dummyy_test, MLPCpreds)

#get default parameter model
MLPC = MLPClassifier(random_state=0).fit(dummyX_train, dummyy_train)
MLPCpreds = MLPC.predict(dummyX_test)
MLPCopt['default'] = accuracy_score(dummyy_test, MLPCpreds)
#append best accuracy score to the results of all models
modelcomparison.iloc[5,1] = MLPCopt[max(MLPCopt, key=MLPCopt.get)]
```
